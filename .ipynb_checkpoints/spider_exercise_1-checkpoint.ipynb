```python
# 4、解析 HTML
from bs4 import BeautifulSoup
soup = BeautifulSoup(req.text, "lxml") #定制为“lxml” 解析器，转化为树形结构文档
#可以使用 find() find_all() 搜索文档
print (soup.title) #通过签名直接获取网页元素内容
```

```python
 print(soup.title.text) #获取文本内容
```

```python
print(soup.find("ul",{"class":"lists"})) 
#通过 find() 查找网页中，class 为 lists 的元素内容
```

```python
lis = soup.find_all("li",{"class":"list-item"})
print (lis[0:5])
# 通过 find_all() 查找网页中，所有 class 为 list-item 的元素内容
```

```python
# 五、抽取内容
for li in soup.find_all("li",{"class":"list-item"}):
    print(li.find("li",{"class":"stitle"}).find("a").text.strip())
```

```python
for li in soup.find_all("li",{"class":"list-item"}):
    print( li.get("data-title"), li.get("data-score"), li.get("data-duration"), li.get("data-star"))
```

```python
output = open("moveies.csv", "w")
for li in soup.find_all("li", {"class":"list-item"}):
    line = "%s\t%s\t%s\n" %( li.get("data-title"), li.get("data-score"), li.get("data-duration"))
    output.write(line)
output.close()
```
